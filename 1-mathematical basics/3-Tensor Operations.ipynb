## The gears of neural networks: tensor operations

```python
import time
import numpy as np
```

### Element-wise operations

```python
def vanilla_add(x, y):
    
    t = x.copy()
    row = x.shape[0]
    column = x.shape[1]
    
    for i in range(row):
        for j in range(column):
            t[i, j] = x[i, j] + y[i, j]
    
    return x
```

```python
def vanilla_relu(x):
    
    t = x.copy()
    row = x.shape[0]
    column = x.shape[1]
    
    for i in range(row):
        for j in range(column):
            t[i, j] = max(x[i, j], 0)
    
    return x
```

```python
x = np.random.random((20, 100))
y = np.random.random((20, 100))
t0 = time.time()
```

### Vanilla Operation

```python
for _ in range(1000):
    
    z = vanilla_add(x, y)
    z = vanilla_relu(z)
    print("Took: {0:.2f} s".format(time.time() - t0))
```

### Built-in Operation

```python
t0 = time.time()

for _ in range(1000):
    
    z = x + y
    z = np.maximum(z, 0.)

    print("Took: {0:.2f} s".format(time.time() - t0))
```

#### Vanilla Broadcasting

```python
def vanilla_add_matrix_and_vector(x, y):

    x = x.copy()
    row = x.shape[0]
    column = x.shape[1]

    for i in range(row):
        for j in range(column):
            x[i, j] += y[j]
    
    return x
```

```python
x = np.random.random((5, 10))
y = np.random.random((10,))
z = vanilla_add_matrix_and_vector(x, y)
z.shape
```

```python
def vanilla_vector_dot(x, y):
    
    z = 0.
    
    for i in range(x.shape[0]):
        z += x[i] * y[i]
    
    return z
```

```python
vanilla_vector_dot(x,y)
```

```python
def vanilla_matrix_vector_dot(x, y):
    
    z = np.zeros(x.shape[0])
    
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            z[i] += x[i, j] * y[j]
    
    return z
```

```python
x = np.array([[1,2],[3,4]])
y = np.array([1,2])
vanilla_matrix_vector_dot(x,y)
```
